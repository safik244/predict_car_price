```python
from sklearn.preprocessing import LabelEncoder
le_fueltype = LabelEncoder()
le_aspiration =LabelEncoder()
le_doornumber = LabelEncoder()
le_carbody = LabelEncoder()
le_enginelocation = LabelEncoder()
le_enginetype = LabelEncoder()
le_fuelsystem = LabelEncoder()


df['fueltype_n'] = le_fueltype.fit_transform(df['fueltype'])
df['aspiration_n'] = le_aspiration.fit_transform(df['aspiration'])
df['doornumbe_n'] = le_doornumber.fit_transform(df['doornumber'])
df['carbody_n'] = le_carbody.fit_transform(df['carbody'])
df['enginelocation_n'] = le_enginelocation.fit_transform(df['enginelocation'])
df['enginetype_n'] = le_enginetype.fit_transform(df['enginetype'])
df['fuelsystem_n'] = le_fuelsystem.fit_transform(df['fuelsystem'])
```

```python
df
```

```python
df.columns
```

```python
upper = df['fueltype_n'].mean() + 3*df['fueltype_n'].std()
lower = df['fueltype_n'].mean()  - 3*df['fueltype_n'].std()
df[(df['fueltype_n']>lower) & (df['fueltype_n']<upper)]
```

```python
x = df[['fueltype_n','aspiration_n','doornumbe_n','carbody_n','enginelocation_n','wheelbase','carlength','carwidth','carheight','curbweight','enginetype_n','enginesize','fuelsystem_n','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']]
y= df.price
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state=20)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[1,0,1,3,0,94.5,165.3,63.8,54.5,1918,3,97,1,3.15,3.29,9.4,69,5200,31,37]])
```

```python
from sklearn.tree import DecisionTreeRegressor
model_1 = DecisionTreeRegressor()
model_1.fit(x_train,y_train)
```

```python
model_1.score(x_test,y_test)
```

```python
model_1.predict([[1,0,0,4,0,95.7,169.7,63.6,59.1,2280,3,92,1,3.05,3.03,9.0,62,4800,31,37]])
```

```python
from sklearn.svm import SVR
model_2 = SVR(kernel="linear",gamma='auto',C=100)
model_2.fit(x_train,y_train)
```

```python
model_2.score(x_test,y_test)
```

```python
model_2.predict([[1,0,1,3,0,94.5,165.3,63.8,54.5,1918,3,97,1,3.15,3.29,9.4,69,5200,31,37]])
```
